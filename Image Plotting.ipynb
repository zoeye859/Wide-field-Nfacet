```python
import numpy as np
from numpy import loadtxt
import scipy
from scipy import constants
import math
import time
import matplotlib.pyplot as plt
%matplotlib notebook
def I_rotation(size,I):
    I_r = np.zeros((size,size))
    for i in range(size):
        for j in range(size):
            I_r[size-1-i,j] = I[j,i] 
    return I_r

def RMS(y):
    return np.sqrt(np.mean(y**2))
```

```python
I_sum = np.loadtxt('I_DFT_900_out6db.csv', delimiter = ',')
I_sum_r = I_rotation(900,I_sum)
fig1, ax1 = plt.subplots()

img1 = ax1.imshow(I_sum_r, cmap='binary', origin = 'lower')
# Where we want the ticks, in pixel locations
ticks = np.array((0,150,300,450,600,750,900))
ticklabelsx = ["{:6.1f}$^\circ$".format(i) for i in (450-ticks)*24/3600.]
ticklabelsy = ["{:6.1f}$^\circ$".format(i) for i in (ticks-450)*24/3600.]
ax1.set_xticks(ticks)
ax1.set_xticklabels(ticklabelsx)
ax1.set_yticks(ticks)
ax1.set_yticklabels(ticklabelsy)
ax1.set_title(r'DFT dirty image')
ax1.set_xlabel('Right Ascension (J2000)')
ax1.set_ylabel('Declination (J2000)')
fig1.colorbar(img1)
fig1.show()
#fig1.savefig('W-Stacking_dirty.png', bbox_inches = 'tight', dpi = 300)
```
